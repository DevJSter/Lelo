```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Get user input for number of data points
num_points = int(input("Enter the number of data points: "))

# Get user input for data points
data_points = []
for i in range(num_points):
    point = input(f"Enter data point {i + 1} as comma-separated values (e.g., x, y): ").split(',')
    data_points.append([float(point[0]), float(point[1])])

# Converting the list of data points to a NumPy array
X = np.array(data_points)

# Get user input for the number of clusters
num_clusters = int(input("Enter the number of clusters: "))

# Initializing KMeans with the number of clusters specified by the user
kmeans = KMeans(n_clusters=num_clusters)

# Fitting the data to the KMeans algorithm
kmeans.fit(X)

# Getting the centroids and labels based on the trained KMeans algorithm
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Visualizing the data points and centroids
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, linewidths=3, color='r')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KMeans Clustering')
plt.show()
```
